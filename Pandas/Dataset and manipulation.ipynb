# **Rating using python list**

```python
import csv
```

```python
def rating_status_calculation(data, industry = None):
  rating = []
  for row in data:
    if row[3]!= 'NULL' and (not industry or row[1] == industry):
      rating.append(float(row[3]))
  min_rating = min(rating)
  max_rating = max(rating)
  avg_rating = sum(rating)/len(rating)
  return min_rating, max_rating, avg_rating
```

```python
def get_industries(data):
  industries =[]
  for row in data:
    if row[1] not in industries:
      industries.append(row[1])
  return industries
```

```python
with open(path, 'r') as file:
    reader = list(csv.reader(file))
    header = reader[0]
    data = reader[1:]
print(header)
print(data[0])
```

```python
get_industries(data)
```

```python
b_min, b_max, b_avg =rating_status_calculation(data, industry='Bollywood')
print(f"Bollywood ratings : min {b_min} , max {b_max} , avg {b_avg}")

h_min, h_max, h_avg = rating_status_calculation(data, industry='Hollywood')
print(f"Hollywood ratings : min {h_min} , max {h_max} , avg {h_avg}")

o_min, o_max, o_avg =rating_status_calculation(data)
print(f"Overall ratings : min {o_min} , max {o_max} , avg {o_avg}")
```
