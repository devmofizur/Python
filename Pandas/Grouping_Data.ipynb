```python
group_object.size()
```

```python
def grouping(df,idx,column):
  if df[column][idx] > 86:
    return 'hot'
  elif df[column][idx] < 50:
    return 'cold'
  else:
    return 'warm'
```

```python
group_object = df.groupby(lambda idx : grouping(df,idx, 'temperature'))
```

```python
for key,data in group_object:
  print("Key : ", key)
  print("Data : ",data)
```

```python
from google.colab import drive
import shutil
import os

drive.mount('/content/drive')

source_path = '/content/drive/MyDrive/Colab Notebooks/ML Learning/movies_data.csv'
destination_path = '/content/movies_data.csv'

if os.path.exists(source_path):
    shutil.copy(source_path, destination_path)
    print(f"Successfully copied '{source_path}' to '{destination_path}'")
else:
    print(f"Error: Source file '{source_path}' not found.")
```

```python
import pandas as pd
df = pd.read_csv(destination_path)
df
```

```python
g = df.groupby('industry')
g.size()
```

```python
for industry,data in g:
  if industry == 'Bollywood':
    print(industry,"\n",data)
```

```python
def grouper(df, idx, col):
  if 1 <= df[col][idx] <= 3.9:
    return 'Poor'
  elif 4 <= df[col][idx] <= 7.9:
    return 'Average'
  elif 8 <= df[col][idx] <= 10:
    return 'Good'
  else:
    return 'others'
```

```python
g = df.groupby(lambda idx: grouper(df,idx,'imdb_rating'))
for key,data in g:
  print('Key : ', key)
  print('Data : \n',data)
```
