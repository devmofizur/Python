```python
# Customer ID, Name
customers = np.array([
    [1, 'Alice'],
    [2, 'Bob'],
    [3, 'Charlie']
])
# Customer ID, purchase amount , purchase date
purchases = np.array([
    [1, 250.0, '2026-04-01'],
    [2, 300.5, '2026-04-02'],
    [1, 120.0, '2026-04-03']
])
```

```python
horizontal_stack = np.hstack((customers,purchases)) # combining two numpy array horizontally. yes it is blind and repeats id
horizontal_stack
```

```python
np.hsplit(horizontal_stack,[2])
```

```python
purchases[:,1:3]
```

```python
vertical_stack = np.vstack((customers,purchases[:,1:3])) # combining two numpy array vertically. yes it is blind and repeats id
vertical_stack
```

```python
np.hsplit(vertical_stack,[1])
```

```python
np.max(horizontal_stack[:,3].astype(np.float64)) # max value from purchase amount converted data type unicode to float64
```

```python
np.argmax(horizontal_stack[:,3].astype(np.float64)) # max value index from purchase amount converted data type unicode to float64
```
